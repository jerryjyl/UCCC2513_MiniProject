## Create custom images
uint8 --> unsigned int, 8-bit

```python
import matplotlib.pyplot as plt
```

```python
# create a black grayscale image

black = np.zeros((200,100), dtype=np.uint8) 
plt.imshow(black, cmap=plt.cm.gray)
plt.title("black img")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# create a white grayscale image

white = np.zeros((200,100)) + 255
white = np.uint8(white)
# white = white.astype("uint8")

plt.imshow(white, cmap=plt.cm.gray, vmin=0, vmax=250)
plt.title("white img")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# random color image

random_color_img = np.random.randint(0, high = 256, size = (150, 250, 3), dtype = np.uint8)

plt.imshow(random_color_img)
plt.title("random color img")
plt.xticks([]), plt.yticks([])
plt.show()
```
